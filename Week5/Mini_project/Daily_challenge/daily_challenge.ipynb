```python
import pandas as pd
import matplotlib.pyplot as plt
salaries_df = pd.read_csv('datascience_salaries.csv')
salaries_df
#Current dataset doesn't contain rows with possible duplicates
```

```python
grouped = salaries_df.groupby('experience_level')
mean_salary = grouped['salary'].mean()
median_salary = grouped['salary'].median()

plt.figure(figsize=(10, 6))
mean_salary.plot(kind='bar', color='skyblue', label='Mean Salary')
median_salary.plot(kind='bar', color='orange', label='Median Salary')

plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.title('Average Salaries by Experience Level')
plt.legend()

plt.tight_layout()
plt.show()
```
