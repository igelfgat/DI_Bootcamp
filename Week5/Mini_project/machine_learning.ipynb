```python
import pandas as pd

mental_health_df = pd.read_csv('survey.csv')
mental_health_df
```

```python
mental_health_df.info()
```

```python
print(mental_health_df.isnull().sum())
```

```python
filtered_df = mental_health_df[(mental_health_df['Age'] >= 16) & (mental_health_df['Age'] <= 99)]
filtered_df
```

```python
#What is the distribution of mental health conditions among different age groups in the tech industry?
age_mental_health = filtered_df.groupby(['Age', 'mental_health_consequence']).size().reset_index(name='count')
age_mental_health_pivot = age_mental_health.pivot(index='Age', columns='mental_health_consequence', values='count').fillna(0)
age_mental_health_pivot.head()
```

```python
import matplotlib.pyplot as plt
age_mental_health_pivot.plot(kind='bar', stacked=True, figsize=(15, 7))
plt.title('Distribution of Mental Health Consequences among Different Age Groups')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Mental Health Consequence')
plt.show()
```

```python
#Something happened with dataset, so I should reload it 
mental_health_df = pd.read_csv('survey.csv')
mental_health_df
```

```python
#How does the frequency of mental health issues vary by gender?
mental_health_df['Gender'] = mental_health_df['Gender'].str.lower()
mental_health_df['Gender'] = mental_health_df['Gender'].replace(['m', 'male', 'men', 'man', 'male '], 'male')
mental_health_df['Gender'] = mental_health_df['Gender'].replace(['f', 'female', 'women', 'woman'], 'female')
mental_health_df['Gender'] = mental_health_df['Gender'].apply(lambda x: 'other' if x not in ['male', 'female'] else x)


print(mental_health_df['Gender'])
```

```python

unique_values = mental_health_df['Gender'].unique()
print(unique_values)

```

```python
grouped_mental_health_df = mental_health_df.groupby(['Gender', 'mental_health_consequence']).size().reset_index(name='count')
mental_mental_health_pivot = grouped_mental_health_df.pivot(index='Gender', columns='mental_health_consequence', values='count').fillna(0)
mental_mental_health_pivot
```

```python
mental_mental_health_pivot.plot(kind='bar', stacked=True, figsize=(15, 7))
plt.title('Frequency of Mental Health Issues by Gender')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.legend(title='Mental Health Consequence')
plt.show()
```

```python
#Identify the countries with the highest and lowest reported rates of mental health issues in the tech industry.
grouped_country_mental_health_df = mental_health_df.groupby(['Country', 'mental_health_consequence']).size().reset_index(name='count')
country_mental_health_pivot = grouped_country_mental_health_df.pivot(index='Country', columns='mental_health_consequence', values='count').fillna(0)
country_mental_health_pivot
```

```python
highest_rate_country = country_mental_health_pivot['Yes'].idxmax()
lowest_rate_country = country_mental_health_pivot['Yes'].idxmin()

print(f"Highest rate country: {highest_rate_country} ({country_mental_health_pivot['Yes'][highest_rate_country]})")
print(f"Lowest rate country: {lowest_rate_country} ({country_mental_health_pivot['Yes'][lowest_rate_country]})")
```
