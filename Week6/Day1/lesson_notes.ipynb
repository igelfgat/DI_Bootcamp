```python
import pandas as pd
# Sample data
data = {
    'Country': ['Canada', 'Canada', 'USA', 'Canada', 'USA', 'Canada', 'USA', 'USA'],
    'City': ['Toronto', 'Vancouver', 'New York', 'Toronto', 'Los Angeles', 'Vancouver', 'New York', 'Los Angeles'],
    'Date': ['2024-07-01', '2024-07-01', '2024-07-01', '2024-07-02', '2024-07-01', '2024-07-02', '2024-07-02', '2024-07-02'],
    'Temperature': [28, 25, 30, 32, 35, 31, 33, 34],
    'Precipitation': [5, 0, 2, 0, 0, 3, 1, 0]
}
# Create DataFrame
df = pd.DataFrame(data)
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
filtered_df = df.loc[('Canada')]
filtered_df = filtered_df[filtered_df['Temperature'] > 30]
print(filtered_df)
```
