```python
import pandas as pd
import numpy as np
# Generating sample data
np.random.seed(42)
customer_data = pd.DataFrame({
    'Age': np.random.randint(18, 70, size=200),
    'Annual Expenditure': np.random.normal(loc=50000, scale=15000, size=200).astype(int)
})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(customer_data['Annual Expenditure'])
plt.title('Histogram')
plt.show()
```

```python
sns.kdeplot(customer_data['Annual Expenditure'], shade=True)
plt.title('KDE of Annual Expenditure')
plt.show()
```

```python
#Univariate Distribution:
sns.histplot(customer_data['Annual Expenditure'], color='green')
plt.title('Univariate Distribution of Annual Expenditure')
plt.show()
```

```python
#Bivariate Distribution:
sns.jointplot(x='Annual Expenditure', y='Age', data=customer_data, kind='scatter')
plt.title('Bivariate Distribution of Total Bill and Tips', pad=70)
plt.show()
```
