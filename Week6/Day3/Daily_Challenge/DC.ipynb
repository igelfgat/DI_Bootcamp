```python
import pandas as pd
df = pd.read_csv("googleplaystore.csv")
# TODO : show only the first 5 results
df.head()
```

```python
# TODO : import sqlalchemy and create a sqlite engine

# TODO : export the dataframe as a table 'playstore' to the sqlite engine
# Import sqlalchemy and create a sqlite engine
from sqlalchemy import create_engine, text

# Create the engine with check_same_thread=False to avoid threading issues
engine = create_engine('sqlite://', echo=False, connect_args={'check_same_thread': False})

# Export the dataframe as a table 'playstore' to the sqlite engine
df.to_sql("playstore", con=engine, if_exists='replace', index=False)

# Obtain a connection from the engine
with engine.connect() as connection:
    # Execute the SQL query using the connection
    result = connection.execute(text("SELECT * FROM playstore"))
    
```

```python
# TODO: Query the database and format the result as a dataframe
query = "SELECT * FROM playstore"
result_df = pd.read_sql(query, engine)
# TODO: Only show the first 5 results / tuples
result_df.head()
```

```python
# TODO : write a query that returns the names of apps (App), their genres (Genres), and their sizes (Size) from the playstore table, 
#where the size of the app is greater than 10 and the genre does not start with 'Art & Design'.
# TODO : create a pandas DataFrame from the results of the SQL query and then displays the first five rows of this DataFrame, 
#with the columns appropriately named according to the SQL query result.
query = "SELECT App, Genres, Size FROM playstore WHERE Size > 10 AND Genres NOT LIKE 'Art & Design%'"
result_df = pd.read_sql(query, engine)
result_df.head()
```
