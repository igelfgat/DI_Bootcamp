```python
import pandas as pd
import io
import requests
#another way to read_csv
url = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/master/inst/extdata/penguins.csv'
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
df
```

```python
# Filter rows with at least one NaN value
rows_with_nan = df[df.isnull().any(axis=1)]
rows_with_nan
```

```python
df.dropna(inplace=True)
```

```python
# Use Pandas to summarize the key statistics of the dataset.
#Identify unique values, ranges, and distribution of data.
df.describe(include='all')
```

```python
# Range of values for numerical columns
numerical_columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
ranges = df[numerical_columns].max() - df[numerical_columns].min()
print(ranges)
```

```python
# Histogram for numerical columns
import matplotlib.pyplot as plt
print("\nHistograms for numerical columns:")
df.hist(bins=30, figsize=(10, 10))
plt.tight_layout()
plt.show()
```

```python
# Box plot for numerical columns
print("\nBox plots for numerical columns:")
df.plot(kind='box', figsize=(10, 6), subplots=True, layout=(2, 3))
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
#Species Distribution: Use a bar chart to show the distribution of penguin species.
species_counts = df['species'].value_counts()
species_counts.plot(kind='bar', color='skyblue')
plt.title('Species Distribution')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
#Flipper Length vs. Body Mass: Create a scatter plot to analyze the relationship between flipper length and body mass, color-coded by species.
import seaborn as sns
sns.lmplot(x='flipper_length_mm', y='body_mass_g', data=df)

plt.title('Flipper Length vs. Body Mass')
plt.xlabel('Flipper length(mm)')
plt.ylabel('Body mass(g)')

plt.show()
```

```python
#Island Distribution: Visualize the count of penguins on each island using a pie chart.
island_counts = df['island'].value_counts()
island_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Island Distribution')
plt.ylabel('')
plt.show()
```

```python
df.head()
```

```python
#Sex Distribution within Species: Create grouped bar charts showing the distribution of sex within each penguin species.
sex_species_counts = df.groupby(['species', 'sex']).size().unstack()
sex_species_counts.plot(kind='bar', stacked=False, color=['red', 'blue'])
plt.title('Sex Distribution within Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python

#Heatmap of Correlations: Plot a heatmap to visualize the correlations between numerical variables.
correlation_matrix = df[numerical_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlations between numerical variables")
plt.show()
```

```python
#Boxplots for Measurements: Generate boxplots for flipper length, body mass, and bill measurements, segmented by species.

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.boxplot(data=df, x='species', y='flipper_length_mm', palette='Set2')
plt.title('Flipper Length by Species')

plt.subplot(2, 2, 2)
sns.boxplot(data=df, x='species', y='body_mass_g', palette='Set2')
plt.title('Body Mass by Species')

plt.subplot(2, 2, 3)
sns.boxplot(data=df, x='species', y='bill_length_mm', palette='Set2')
plt.title('Bill Length by Species')

plt.subplot(2, 2, 4)
sns.boxplot(data=df, x='species', y='bill_depth_mm', palette='Set2')
plt.title('Bill Depth by Species')

plt.tight_layout()
plt.show()
```

```python
#Histograms of Measurements: Create histograms to show the distribution of key measurements (bill length, bill depth, flipper length).
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
sns.histplot(data=df, x='bill_length_mm', kde=True, color='skyblue')
plt.title('Distribution of Bill Length')

plt.subplot(3, 1, 2)
sns.histplot(data=df, x='bill_depth_mm', kde=True, color='purple')
plt.title('Distribution of Bill Depth')

plt.subplot(3, 1, 3)
sns.histplot(data=df, x='flipper_length_mm', kde=True, color='lightgreen')
plt.title('Distribution of Flipper Length')

plt.tight_layout()
plt.show()
```

Insightful Visualizations:

Combine multiple plots into a single figure (using subplots) to compare different aspects of the data.
Use Seaborn’s pairplot to visualize pairwise relationships in the dataset.

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

species_counts = df['species'].value_counts()
species_counts.plot(kind='bar', color='skyblue', ax=axs[0, 0])
axs[0, 0].set_title('Species Distribution')
axs[0, 0].set_xlabel('Species')
axs[0, 0].set_ylabel('Count')

sns.scatterplot(data=df, x='flipper_length_mm', y='body_mass_g', hue='species', palette='viridis', ax=axs[0, 1])
axs[0, 1].set_title('Flipper Length vs. Body Mass')
axs[0, 1].set_xlabel('Flipper Length (mm)')
axs[0, 1].set_ylabel('Body Mass (g)')

island_counts = df['island'].value_counts()
island_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'), ax=axs[1, 0])
axs[1, 0].set_title('Island Distribution')
axs[1, 0].set_ylabel('')

sex_species_counts = df.groupby(['species', 'sex']).size().unstack()
sex_species_counts.plot(kind='bar', stacked=False, color=['red', 'blue'], ax=axs[1, 1])
axs[1, 1].set_title('Sex Distribution within Species')
axs[1, 1].set_xlabel('Species')
axs[1, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()

```

```python
# Use Seaborn’s pairplot to visualize pairwise relationships in the dataset.
sns.pairplot(df, hue='species', palette='viridis')
plt.show()
```
