Instructions

Given a sample dataset, calculate the mean, median, variance, and standard deviation using SciPy.
Sample dataset in code:

```python
data = [12, 15, 13, 12, 18, 20, 22, 21]
mean_scipy = stats.tmean(data)
median_scipy = np.median(data)  # SciPy does not have a separate median function
variance_scipy = stats.tvar(data)
std_dev_scipy = stats.tstd(data)
mean_scipy, median_scipy, variance_scipy, std_dev_scipy
```

 Exercise 3 : Understanding Distributions

Instructions

Generate a normal distribution using SciPy with a mean of 50 and a standard deviation of 10. Plot this distribution.

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 50
std_dev = 10

# Generate the normal distribution
# Using linspace to cover the range from mean - 4*std_dev to mean + 4*std_dev
data = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)
pdf = norm.pdf(data, mean, std_dev)

plt.figure(figsize=(10, 6))
plt.plot(data, pdf, label=f'Normal Distribution (mean={mean}, std={std_dev})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

 Exercise 4 : T-Test Application

Instructions

Perform a T-test on two sets of randomly generated data. Hint: Use numpy.random.normal() to generate the datasets.

```python
mean1, std_dev1, size1 = 50, 10, 100 
mean2, std_dev2, size2 = 55, 12, 100 

data1 = np.random.normal(loc=mean1, scale=std_dev1, size=size1)
data2 = np.random.normal(loc=mean2, scale=std_dev2, size=size2)

t_stat, p_value = stats.ttest_ind(data1, data2)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("The difference between the two datasets is statistically significant.")
else:
    print("There is no significant difference between the two datasets.")
```
