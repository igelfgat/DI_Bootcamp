ðŸŒŸ Exercise 1 : Linear Regression Analysis

Instructions

Objective: Apply linear regression to a dataset and interpret the results.

Task: Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.

Dataset:

house_sizes: [50, 70, 80, 100, 120] (in square meters)
house_prices: [150,000, 200,000, 210,000, 250,000, 280,000] (in currency units)
Questions:

What is the slope and intercept of the regression line?
Predict the price of a house that is 90 square meters.
Interpret the meaning of the slope in the context of housing prices.

```python
import numpy as np
from scipy import stats
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices =np.array([150000, 200000, 210000, 250000, 280000])
# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

print("Slope:", slope)
print("Intercept:", intercept)
```

```python
import matplotlib.pyplot as plt
# Generate regression line values
y_pred = slope * house_sizes + intercept # y = mx + b ## m-slope, b-intercept
print(y_pred)
# Plot data points and regression line
plt.figure(figsize=(10, 6))
plt.scatter(house_sizes, house_prices, color='blue', label='Data Points')
plt.plot(house_sizes, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression Analysis')
plt.xlabel('House sizes')
plt.ylabel('House prices')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Predict the price of a house that is 90 square meters.
predicted_price_90 = slope * 90 + intercept
print("Predicted price for 90 square meters:", predicted_price_90)
```

Interpret the meaning of the slope in the context of housing prices
The slope represents the change in house price for each additional square meter of house size. For instance, if the slope is 2000, it means that for every additional square meter, the house price increases by 2000 currency units.

ðŸŒŸ Exercise 2 : Understanding ANOVA

Instructions

Objective: Test understanding of ANOVA and its application.

Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.

Dataset:

fertilizer_1: [5, 6, 7, 6, 5]
fertilizer_2: [7, 8, 7, 9, 8]
fertilizer_3: [4, 5, 4, 3, 4]
Questions:

Perform an ANOVA test on the given data. What are the F-value and P-value?
Based on the P-value, do the fertilizers have significantly different effects on plant growth?
Explain what would happen if the P-value were greater than 0.05.

```python
fertilizer_1 = np.array([5, 6, 7, 6, 5])
fertilizer_2 = np.array([7, 8, 7, 9, 8])
fertilizer_3 = np.array([4, 5, 4, 3, 4])
# Performing one-way ANOVA
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value)
print("P-value:", p_value)
```

The F-value of 28.526 indicates a large ratio of the variance between the group means to the variance within the groups. This suggests that the differences between the means of the groups are substantial relative to the variability within the groups.
The P-value of 2.754e-05 (0.00002754) is significantly less than the common alpha level of 0.05. This means there is a very low probability that the observed differences between the groups are due to random chance.


Significant Differences: Because the P-value is much less than 0.05, we reject the null hypothesis. This indicates that there are significant differences between the effects of the three fertilizers on plant growth.

Explain what would happen if the P-value were greater than 0.05:
This would mean that any differences in plant growth between the fertilizers are not statistically significant and could be due to random variation

ðŸŒŸ Exercise 3 : Probability Distributions

Instructions

Work with a binomial distribution: calculate probabilities for different numbers of successes.
Example: Calculating the probability of getting exactly 5 heads in 10 coin flips.

```python
n = 10  # number of trials
p = 0.5  # probability of success
k = 5  # number of successes

probability = stats.binom.pmf(k, n, p)

print(f"The probability of getting exactly {k} heads in {n} coin flips is {probability:.4f}")
```

```python
k_values = np.arange(0, n+1)

# Calculate the probabilities for each number of successes
probabilities = stats.binom.pmf(k_values, n, p)

# Plotting the probability distribution
plt.figure(figsize=(10, 6))
plt.bar(k_values, probabilities, color='blue', alpha=0.7)
plt.xlabel('Number of Heads (Successes)')
plt.ylabel('Probability')
plt.title('Binomial Distribution: Probability of Getting Heads in 10 Coin Flips')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

 Exercise 4 : Correlation Coefficients

Instructions

Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.
Example in code: data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})

```python
import pandas as pd
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
```

```python
from scipy.stats import pearsonr, spearmanr
# Calculate Pearson correlation
pearson_corr, _ = pearsonr(data['age'], data['income'])

# Calculate Spearman correlation
spearman_corr, _ = spearmanr(data['age'], data['income'])

print(f"Pearson correlation coefficient: {pearson_corr:.4f}")
print(f"Spearman correlation coefficient: {spearman_corr:.4f}")
```
