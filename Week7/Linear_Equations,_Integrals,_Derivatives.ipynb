### Overview of Key Concepts

1. **Equations**:
   - **Definition**: Equations represent relationships between variables. Solving equations means finding the values of the variables that make the equation true.
   - **Use in Data Analysis**: Equations can model relationships between different data points, helping to predict future trends and optimize processes.

2. **Integrals**:
   - **Definition**: Integration is the process of finding the area under a curve, which represents the accumulation of quantities over time or space.
   - **Use in Data Analysis**: Integrals help in calculating total quantities, such as total sales over a period, total engagement, or accumulated profits.
   - **Manual Computation**: To integrate a function like \( f(x) = x^2 \), we increase the exponent by 1 and divide by the new exponent. The integral of \( x^n ) is x^(n+1)/n+1.

3. **Derivatives**:
   - **Definition**: Differentiation is the process of finding the rate of change of a quantity. It tells us how a function changes as its input changes.
   - **Use in Data Analysis**: Derivatives are used to find trends, optimize functions, and determine maximum and minimum values of data-related functions.
   - **Manual Computation**: To differentiate a function like \( f(x) = x^2 \), we bring the exponent down as a coefficient and subtract 1 from the exponent. The derivative of \( x^n \) is \( nx^{n-1} \).

### Manual Computation Examples

#### Solving a System of Equations Manually (Row Manipulation)

Consider the system of equations:

\[
\begin{cases}
2x + y + z = 3 \\
x + 3y + z = 4 \\
x + y + 4z = 5
\end{cases}
\]

Convert this system into matrix form  Ax=b:

\
A = \begin{pmatrix}
2 & 1 & 1 \\
1 & 3 & 1 \\
1 & 1 & 4
\end{pmatrix}x = \begin{pmatrix}
x \\
y \\
z
\end{pmatrix}
b = \begin{pmatrix}
3 \\
4 \\
5
\end{pmatrix}


**Step-by-Step Row Manipulation**:

1. **Initial Matrix**:
\[
\left[\begin{array}{ccc|c}
2 & 1 & 1 & 3 \\
1 & 3 & 1 & 4 \\
1 & 1 & 4 & 5
\end{array}\right]
\]

2. **Make the first element of the first row 1** (by dividing the first row by 2):
\[
\left[\begin{array}{ccc|c}
1 & \frac{1}{2} & \frac{1}{2} & \frac{3}{2} \\
1 & 3 & 1 & 4 \\
1 & 1 & 4 & 5
\end{array}\right]
\]

3. **Subtract the first row from the second and third rows to make their first elements 0**:
\[
\left[\begin{array}{ccc|c}
1 & \frac{1}{2} & \frac{1}{2} & \frac{3}{2} \\
0 & \frac{5}{2} & \frac{1}{2} & \frac{5}{2} \\
0 & \frac{1}{2} & \frac{7}{2} & \frac{7}{2}
\end{array}\right]
\]

4. **Make the second element of the second row 1** (by dividing the second row by \( \frac{5}{2} \)):
\[
\left[\begin{array}{ccc|c}
1 & \frac{1}{2} & \frac{1}{2} & \frac{3}{2} \\
0 & 1 & \frac{1}{5} & 1 \\
0 & \frac{1}{2} & \frac{7}{2} & \frac{7}{2}
\end{array}\right]
\]

5. **Subtract the second row from the first and third rows to make their second elements 0**:
\[
\left[\begin{array}{ccc|c}
1 & 0 & \frac{2}{5} & \frac{1}{2} \\
0 & 1 & \frac{1}{5} & 1 \\
0 & 0 & \frac{33}{10} & 3
\end{array}\right]
\]

6. **Make the third element of the third row 1** (by dividing the third row by \( \frac{33}{10} \)):
\[
\left[\begin{array}{ccc|c}
1 & 0 & \frac{2}{5} & \frac{1}{2} \\
0 & 1 & \frac{1}{5} & 1 \\
0 & 0 & 1 & \frac{30}{33}
\end{array}\right]
\]

7. **Subtract appropriate multiples of the third row from the first and second rows to make their third elements 0**:
\[
\left[\begin{array}{ccc|c}
1 & 0 & 0 & \frac{1}{2} - \frac{2}{5} \cdot \frac{30}{33} \\
0 & 1 & 0 & 1 - \frac{1}{5} \cdot \frac{30}{33} \\
0 & 0 & 1 & \frac{30}{33}
\end{array}\right]
\]

8. **Simplify the fractions**:
\[
\left[\begin{array}{ccc|c}
1 & 0 & 0 & 0.647 \\
0 & 1 & 0 & 0.824 \\
0 & 0 & 1 & 0.882
\end{array}\right]
\]

Thus, the solution is:
\[
\mathbf{x} = \begin{pmatrix}
x \\
y \\
z
\end{pmatrix} = \begin{pmatrix}
0.647 \\
0.824 \\
0.882
\end{pmatrix}
\]

**Code Implementation**:

```python
import numpy as np

# Define the matrix A
A = np.array([
    [2, 1, 1],
    [1, 3, 1],
    [1, 1, 4]
])

# Define the vector b
b = np.array([3, 4, 5])

# Solve the linear system A * x = b
x = np.linalg.solve(A, b)

x
```

Running this code will yield the solution:
x = \begin{pmatrix} 0.647 \\ 0.824 \\ 0.882 \end{pmatrix} \]

### Real-World Scenario for Integration

**Scenario**: A company wants to calculate the total revenue generated over a month given a continuous revenue function.

**Example Question**: Calculate the total revenue over a month (30 days) if the daily revenue function is given by \( R(t) = 500e^{-0.05t} \) dollars per day. The revenue function models the decay in daily revenue due to market saturation or reduced customer interest over time.


\]image.png

Using Python to compute:
```python
from scipy.integrate import quad
import numpy as np

# Define the revenue function
def revenue(t):
    return 500 * np.exp(-0.05 * t)

# Integrate the function from 0 to 30
total_revenue, _ = quad(revenue, 0, 30)

total_revenue
```

Running this code will yield the total revenue over 30 days.

### Real-World Scenario for Differentiation

**Scenario**: An analyst wants to determine the rate of change of sales at a particular time to understand the trend of sales growth or decline.

**Example Question**: Given the sales function \( S(t) = 200t^2 - 50t + 1000 \), find the rate of change of sales at \( t = 10 \) days. This sales function could represent a quadratic growth model where sales initially grow but then decelerate due to market saturation or other factors.

#### Solution:

To find the rate of change at \( t = 10 \), we need to differentiate the sales function and then evaluate it at \( t = 10 \).

image.png
Using Python:
```python
import sympy as sp

# Define the variable and function
t = sp.symbols('t')
S = 200 * t**2 - 50 * t + 1000

# Compute the derivative
dS_dt = sp.diff(S, t)

# Evaluate the derivative at t = 10
rate_of_change = dS_dt.evalf(subs={t: 10})

rate_of_change
```

Running this code will yield the rate of change of sales at \( t = 10 \) days.

### Real-World Scenario for Solving Systems of Linear Equations

#### Scenario:
A data analyst at a retail company is tasked with optimizing inventory levels for three products to maximize profit. The company has limited storage space and a budget constraint. The analyst needs to determine the optimal quantity of each product to stock.

The constraints are as follows:

1. The total storage space used by the products should not exceed the available space.
2. The total cost of the products should not exceed the budget.
3. The profit from selling the products should be maximized.

#### Example Question:
You are a data analyst at a retail company. You need to determine the optimal quantities of three products (A, B, and C) to stock in order to maximize profit, given the following constraints:

1. Each unit of Product A takes 2 cubic meters of space, Product B takes 1 cubic meter, and Product C takes 1 cubic meter. The total storage space available is 100 cubic meters.
2. The cost per unit of Product A is $50, Product B is $30, and Product C is $20. The total budget available is $1500.
3. The profit per unit of Product A is $100, Product B is $70, and Product C is $50. The total profit should be maximized.

Formulate the problem as a system of linear equations and solve for the optimal quantities of Products A, B, and C to stock.

#### Solution:
Given the constraints, we have the following system of equations:

\[
\begin{cases}
2x + y + z = 100 \quad \text{(Storage space constraint)} \\
50x + 30y + 20z = 1500 \quad \text{(Budget constraint)} \\
100x + 70y + 50z = \text{maximize} \quad \text{(Profit maximization)}
\end{cases}
\]

However, to use a system of linear equations for solving the exact quantities, we need a method like linear programming. Here, we'll focus on the constraints and solve the equations for feasible quantities.

```python
import numpy as np

# Define the matrix A (coefficients of x, y, z)
A = np.array([
    [2, 1, 1],
    [50, 30, 20]
])

# Define the vector b (right-hand side values)
b = np.array([100, 1500])

# Solve the linear system A * x = b
x = np.linalg.lstsq(A, b, rcond=None)[0]

x
```

### Real-World Scenario for Integrals and Derivatives

#### Scenario:
A data analyst is working for an online streaming service to analyze the user engagement data. The goal is to understand how the engagement varies over time and to predict future trends.


image.png


Using python:
```python
from scipy.integrate import quad
import numpy as np

# Define the engagement function
def engagement(t):
    return 10 * np.exp(-0.1 * t) * np.sin(t)

# Integrate the function from 0 to 24
total_engagement, _ = quad(engagement, 0, 24)

total_engagement
```


image.png

(dE/dt means the derivative of E by the time. Same way we say derivative of function y by x is seen as dy/dx)

Using code:
```python
import sympy as sp

# Define the variable and function
t = sp.symbols('t')
E = 10 * sp.exp(-0.1 * t) * sp.sin(t)

# Compute the derivative
dE_dt = sp.diff(E, t)

# Evaluate the derivative at t = 5
rate_of_change = dE_dt.evalf(subs={t: 5})

rate_of_change
```
