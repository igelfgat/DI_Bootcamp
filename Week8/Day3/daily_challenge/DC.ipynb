```python
import re
input_file_path = 'globalterrorismdb_0718dist.csv'

output_file_path = 'globalterrorismdb_0718dist_cleaned.csv'


with open(input_file_path, 'r', encoding='utf-8') as file:
    content = file.read()

# remove unusual line terminators
cleaned_content = re.sub(r'\u2028|\u2029', '', content)

# write the cleaned content back to a new file
with open(output_file_path, 'w', encoding='utf-8') as file:
    file.write(cleaned_content)
```
